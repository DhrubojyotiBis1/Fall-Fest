# Data preparation

```python
X=(df.iloc[:, [1,23,13,8,35,16,20,18]])
#Creating a dataframe with some of the important parameters that the user can be able to input
X
#checking the dataframe
```

```python
pd.unique(X['Influenza B'])
#Variables like Influenza B contains categorical values 
```

```python
X.dtypes
#Checking the datatyoes of all the variables in dataframe X
```

```python
obj_df = X.select_dtypes(include=['object']).copy()
#Creating a dataframe bby copying those variables which are of object type 

obj_df.head()
#Checking the first five rows of obj_df dataframe
```

```python
obj_df[obj_df.isnull().any(axis=1)]
#Checking the positions where null values are present 
#A lot of values are null. Hence, we have to fill the appropriately 
```

```python
cleanup_nums = {"Influenza B":     {"detected": 1, "not_detected": 0},
                "Inf A H1N1 2009": {"detected": 1, "not_detected": 0} }
#Creating a dictionary to map categorical values to integers 
```

```python
obj_df = obj_df.replace(cleanup_nums)
#Using the repace method to convert categorical values to factors

obj_df.head()
```

```python
X_=(df.iloc[:, [1,13,8,16,20,18]])
#X_ is a dataframe created after excluding the object type variables of X so that the obj_df dataframe can be appended

x= pd.concat([X_, obj_df], axis=1)
```

```python
x
#Final dataframe of independent variable to be used in the model
```

```python
from sklearn.impute import KNNImputer
#To fill the null/missing values, KNN imputer is being used. Hence, importing it from Scikit learn  
```

```python
imputer = KNNImputer(n_neighbors=2)
#Creating an instance of the imputer
```

```python
x=imputer.fit_transform(x)
#Using fit and transform method in the imputer
```

```python
x
#Independent variable array x after imputation
```

```python
y=(df.iloc[:, [2]])
#The dependent variable which contains the result of patients' COVID test 
```

```python
y
```

```python
y[y=='negative']=0
y[y=='positive']=1
#Converting the categorical values to integers
```

```python
y
```

```python
y=np.ravel(y)
```

```python
y.shape
```

# Splitting the data into train and test set

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=0)
#2/3 part of dataset is used for training and the remaining for testing
```

# Training and testing the data 

```python
params = {'n_estimators': 500,
          'max_depth': 4,
          'min_samples_split': 5,
          'learning_rate': 0.01,
          'loss': 'ls'}
```

```python
from sklearn import ensemble
#Importing the model from Sklearn module

reg = ensemble.GradientBoostingRegressor(**params)
#Creating regressor variable 'reg'
```

```python
reg.fit(x_train, y_train)
#training the model 
```

```python
from sklearn.metrics import mean_squared_error
#To calculate the mean squared error

mse = mean_squared_error(y_test, reg.predict(x_test))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```

```python
# evaluate gradient boosting ensemble for regression
from numpy import mean
from numpy import std
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# define the evaluation procedure
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
```

```python
# evaluate the model
n_scores = cross_val_score(reg, x, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
```

```python
# report performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
